# one sample t-test

# imported library and explanation


```python
import numpy as np
from scipy import stats

# --- 1. Generate Data ---
# National average weight (hypothesized population mean)
national_avg_weight = 180 # lbs

# Generate a sample of 50 male student weights
# We create a sample with a mean slightly higher than 180 (e.g., 185) 
# and a standard deviation of 15 to simulate a real-world sample
sample_weights = np.random.normal(loc=185, scale=15, size=50)

print(f"Hypothesized Population Mean (H0): {national_avg_weight} lbs")
print(f"Sample Size: {len(sample_weights)}")
print(f"Sample Mean: {np.mean(sample_weights):.2f} lbs\n")


# --- 2. State Hypotheses ---
# H0 (Null Hypothesis): The true mean weight of the sample IS 180 lbs.
# Ha (Alternative Hypothesis): The true mean weight of the sample IS NOT 180 lbs.


# --- 3. Run the t-test ---
# The ttest_1samp function returns the t-statistic and the two-tailed p-value.
t_statistic, p_value = stats.ttest_1samp(a=sample_weights, popmean=national_avg_weight)

print(f"T-statistic: {t_statistic:.4f}")
print(f"P-value: {p_value:.4f}\n")


# --- 4. Analyze the Results ---
# We use a standard significance level (alpha) of 0.05.
alpha = 0.05

print("-" * 30)
print(f"Significance Level (Alpha): {alpha}")

if p_value < alpha:
    print("Result: Reject the Null Hypothesis (H0)")
    print(f"Conclusion: The average weight of the sample is statistically DIFFERENT from the national average of {national_avg_weight} lbs.")
else:
    print("Result: Fail to reject the Null Hypothesis (H0)")
    print(f"Conclusion: There is not enough evidence to say the sample mean weight is different from the national average of {national_avg_weight} lbs.")

```
