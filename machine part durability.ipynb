# Step-1

```python
import numpy as np
from scipy import stats
```

# step-2

```python
np.random.seed(10)

supplier_A = np.random.normal(1000, 50, 30)
supplier_B = np.random.normal(1020, 50, 30)
supplier_C = np.random.normal(1100, 50, 30)
supplier_D = np.random.normal(1010, 50, 30)
```

# step-3

```python
f_statistic, p_value = stats.f_oneway(
    supplier_A,
    supplier_B,
    supplier_C,
    supplier_D
)
```

# step-4

```python
print("F-Statistic:", f_statistic)
print("P-Value:", p_value)
if p_value < 0.05:
    print("Conclusion: Average durability differs among suppliers.")
else:
    print("Conclusion: No significant difference in average durability.")
```
